# 🌐 Web Scraping

## 📌 Definition
Web scraping is the process of extracting data from websites automatically using scripts or tools.

---

## 🔄 Data Workflow

### 1. Problem Statement
- Identify the problem based on client requirements
- Define what data is needed

### 2. Data Collection
- APIs
- Datasets (e.g., Kaggle)
- Web scraping

### 3. Data Cleaning
- Handle missing values
- Remove duplicates
- Treat outliers
- Fix data formats

### 4. Data Analysis (EDA)
- Understand patterns and trends
- Extract meaningful insights

### 5. Data Visualization
- Represent data using graphs
- Make insights easy to understand

### 6. Conclusion
- Present findings in a storytelling format
- Focus on key insights

# 🧰 Web Scraping Libraries

Web scraping involves extracting data from websites using Python tools.

---

## 📌 1. Regular Expressions (re)
- Used for pattern matching in text
- Helps extract specific data from raw HTML or strings
- Useful for cleaning and filtering data

---

## 📌 2. Requests
- Used to send HTTP requests to websites
- Fetches HTML content of a page
- Simple and lightweight

Example:
- Get webpage source code

---

## 📌 3. Beautiful Soup
- Parses and navigates HTML content
- Works well with static websites
- Extracts data from HTML structure easily

---

## 📌 4. Selenium
- Automates real browser actions
- Used for dynamic websites (JavaScript-loaded content)
- Can click buttons, scroll pages, fill forms

---

## ⚖️ Beautiful Soup vs Selenium

- Beautiful Soup → Fast, lightweight, best for static pages
- Selenium → Powerful, slower, best for dynamic pages

👉 Best practice: Use Beautiful Soup + Requests for most cases, Selenium only when necessary
