```python
### User-defined Function to clean and preprocess the text
def text_clean(x,stem_lem = "stem"):
    ## converting into lower case
    x=x.lower()
    ## replacing html tags
    re.sub("<.*?>","",x)
    ## replace urls
    re.sub("http[s]?://.+?\S+",'',x)
    ## Replacing emoji
    x = emoji.demojize(x)
    x = re.sub(":.*?:","",x)
    ## Replacing special symbols
    x = re.sub("[^a-zA-Z0-9\s_]","",x)
    ## Spell corrector
    blob = TextBlob(x)
    x = str(blob.correct())
    ## Remove stopwords
    z = []
    for word in x.split():
        if word not in stopwords.words("english"):
            z.append(word)
   
    ## Stemming or Lemmatization
    if stem_lem == "stem":
        final = [snbstemmer.stem(ele) for ele in z]
    else:
        final = [lem.lemmatize(ele) for ele in z]
    return ' '.join(final)
```
